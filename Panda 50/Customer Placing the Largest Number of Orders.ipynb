```python
# Source: https://leetcode.com/problems/customer-placing-the-largest-number-of-orders/solutions/6626139/simple-best-solution-by-iqbaldiit-j6ff/
'''
	Table: Orders

	+-----------------+----------+
	| Column Name     | Type     |
	+-----------------+----------+
	| order_number    | int      |
	| customer_number | int      |
	+-----------------+----------+
	order_number is the primary key (column with unique values) for this table.
	This table contains information about the order ID and the customer ID. 

	Write a solution to find the customer_number for the customer who has 
	placed the largest number of orders.
	The test cases are generated so that exactly one customer will have placed 
	more orders than any other customer.

	The result format is in the following example.
	
	Example 1:

	Input: 
	Orders table:
	+--------------+-----------------+
	| order_number | customer_number |
	+--------------+-----------------+
	| 1            | 1               |
	| 2            | 2               |
	| 3            | 3               |
	| 4            | 3               |
	+--------------+-----------------+

| order_number | customer_number |
| ------------ | --------------- |

	Output: 
	+-----------------+
	| customer_number |
	+-----------------+
	| 3               |
	+-----------------+
	Explanation: 
	The customer with number 3 has two orders, which is greater than either customer 1 or 2 because each of them only has one order. 
	So the result is customer_number 3.
'''
import pandas as pd

def largest_orders(orders: pd.DataFrame) -> pd.DataFrame:
    # Count the number of orders placed by each customer
	order_counts = orders["customer_number"].value_counts().reset_index()
	order_counts.columns = ["customer_number", "order_count"]

    # Handle empty DataFrame gracefully
	if order_counts.empty:
		return pd.DataFrame({"customer_number": []})


	# Find the customer with the largest number of orders
	max_orders_customer = order_counts.loc[order_counts["order_count"].idxmax(), "customer_number"]

	# Create the result DataFrame
	result_df = pd.DataFrame({"customer_number": [max_orders_customer]})
	
	return result_df
```

```python
# Sample data for the Orders table
data = {
    "order_number": [1, 2, 3, 4],
    "customer_number": [1, 2, 3, 3]
}

# Create the Orders DataFrame
orders_df = pd.DataFrame(data)
# invoke function
largest_orders(orders_df)
```
